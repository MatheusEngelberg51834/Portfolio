```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pylab as pl
import matplotlib.gridspec as gridspec

colors = {0: 'blue', 1:'red'}
################################
cp_p = {}

for i,j  in zip(df['cp'].value_counts(), df['cp'].keys()):
    cp_p[j] = round(i / 303 * 100, 2)
################################
cp_c = df['cp'].value_counts().sort_index()
cp_0 = df['output'][df['cp'] == 0].value_counts()[1]
cp_1 = df['output'][df['cp'] == 1].value_counts()[1]
cp_2 = df['output'][df['cp'] == 2].value_counts()[1]
cp_3 = df['output'][df['cp'] == 3].value_counts()[1]
###############################
thac_mean = df['thalachh'].mean()
thac_c = [df['thalachh'][df['thalachh'] > thac_mean].shape[0], df['thalachh'][df['thalachh'] <= thac_mean].shape[0]]
thac_below_mean_output = df['output'][df['thalachh'] <= thac_mean].value_counts().sort_index()[1]
thac_above_mean_output = df['output'][df['thalachh'] > thac_mean].value_counts().sort_index()[1]
thac_mean_std = [df['thalachh'][df['thalachh'] > thac_mean].std(), df['thalachh'][df['thalachh'] <= thac_mean].std()]
#################################
slp_c = df['slp'].value_counts().sort_index()
slp_0 = df['output'][df['slp'] == 0].value_counts()[1]
slp_1 = df['output'][df['slp'] == 1].value_counts()[1]
slp_2 = df['output'][df['slp'] == 2].value_counts()[1]
#################################
caa_c = df['caa'].value_counts().sort_index()
caa_0 = df['output'][df['caa'] == 0].value_counts()[1]
caa_1 = df['output'][df['caa'] == 1].value_counts()[1]
caa_2 = df['output'][df['caa'] == 2].value_counts()[1]
caa_3 = df['output'][df['caa'] == 3].value_counts()[1]
caa_4 = df['output'][df['caa'] == 4].value_counts()[1]
#################################
exng_c = df['exng'].value_counts().sort_index()
exng_0 = df['output'][df['exng'] == 0].value_counts()[1]
exng_1 = df['output'][df['exng'] == 1].value_counts()[1]
#################################
zero_thac = df['thalachh'][df['output'] == 0]
one_thac = df['thalachh'][df['output'] == 1]
zero_peak = df['oldpeak'][df['output'] == 0]
one_peak = df['oldpeak'][df['output'] == 1]
zero_output = df['output'][df['output'] == 0]
one_output = df['output'][df['output'] == 1]
#################################
zero_age = df['age'][df['output'] == 0]
one_age = df['age'][df['output'] == 1]
#################################
zero_chol = df['chol'][df['output'] == 0]
one_chol = df['chol'][df['output'] == 1]
#################################
zero_trtbps = df['trtbps'][df['output'] == 0]
one_trtbps = df['trtbps'][df['output'] == 1]
```

```python
gs = gridspec.GridSpec(2, 2,)

pl.figure(figsize=(12, 10))

ax = pl.subplot(gs[0, :])
pl.scatter(zero_thac, zero_peak, c=zero_output.map(colors), alpha=0.5, label='Output = 0')
pl.scatter(one_thac, one_peak, c=one_output.map(colors), alpha=0.5,label='Output = 1')
ax.set_title('Maximum heart rate achieved / Old peak / Output')
ax.legend()

z = np.polyfit(zero_thac, zero_peak, 1)
p = np.poly1d(z)

ax = pl.subplot(gs[1, 0])
pl.scatter(zero_thac, zero_peak, c=zero_output.map(colors), alpha=0.5)
pl.plot(zero_thac, p(zero_thac), 'k--', alpha=0.7)
ax.set_title('Where output = 0')

z = np.polyfit(one_thac, one_peak, 1)
p = np.poly1d(z)

ax = pl.subplot(gs[1, 1])
pl.scatter(one_thac, one_peak, c=one_output.map(colors), alpha=0.5)
pl.plot(one_thac, p(one_thac), 'k--', alpha=0.7)
ax.set_title('Where output = 1')

pl.show()

print('Old peak mean where output = 0: {}'.format(round(zero_peak.mean(), 2)))
print('Old peak mean where output = 1: {}'.format(round(one_peak.mean(), 2)))
print('Old peak std where output = 0: {}'.format(round(zero_peak.std(), 2)))
print('Old peak std where output = 1: {}'.format(round(one_peak.std(), 2)))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import neighbors
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import GridSearchCV

log = LogisticRegression()
tree = tree.DecisionTreeClassifier()
ada = AdaBoostClassifier()
grad = GradientBoostingClassifier()
knn = neighbors.KNeighborsClassifier()
rand = RandomForestClassifier()
```

```python
x_features = ['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh', 'exng', 'oldpeak', 'slp', 'caa', 'thall']
```

```python
log_params = {
    'penalty': ['l2'],
    'tol': [0.0001, 0.001, 0.01, 0.1, 1, 10],
    'C': [0.0001, 0.001, 0.01, 0.1, 1, 10],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
}

log_clf = GridSearchCV(log, log_params, verbose=3, n_jobs=4)

log_clf.fit(df[x_features],df['output'])

print('The best estimator was {}, with {:.3f}% accuracy'.format(log_clf.best_estimator_, (log_clf.best_score_ * 100)))
```

```python
tree_params = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'min_samples_split': [0.1, 1, 2, 5, 10, 50, 100],
    'min_samples_leaf': [0.1, 1, 2, 5, 10, 50, 100]
}

tree_clf = GridSearchCV(tree, tree_params, verbose=1, n_jobs=4, error_score=0)

tree_clf.fit(df[x_features],df['output'])

print('The best estimator was {}, with {:.3f}% accuracy'.format(tree_clf.best_estimator_, (tree_clf.best_score_ * 100)))
```

```python
ada_params = {
    'n_estimators': [10, 20, 50, 100, 500],
    'learning_rate': [0.001, 0.1, 0.2, 0.5, 1, 5, 10],
    'algorithm': ['SAMME', 'SAMME.R'],
}

ada_clf = GridSearchCV(ada, ada_params, verbose=1, n_jobs=4, error_score=0)

ada_clf.fit(df[x_features],df['output'])

print('The best estimator was {}, with {:.3f}% accuracy'.format(ada_clf.best_estimator_, (ada_clf.best_score_ * 100)))
```

```python
grad_params = {
    'loss': ['deviance', 'exponential'],
    'learning_rate': [0.001, 0.1, 0.2, 0.5, 1, 5, 10],
    'n_estimators': [10, 20, 50, 100, 500],
    'criterion': ['friedman_mse', 'mse', 'mae'],
}

grad_clf = GridSearchCV(grad, grad_params, verbose=1, n_jobs=4, error_score=0)

grad_clf.fit(df[x_features],df['output'])

print('The best estimator was {}, with {:.3f}% accuracy'.format(grad_clf.best_estimator_, (grad_clf.best_score_ * 100)))
```

```python
knn_params = {
    'n_neighbors': [1, 2, 3 , 4, 5, 10, 20, 25, 50, 100],
    'weights': ['uniform', 'distance'],
    'p': [2, 3],
    'leaf_size': [1, 2, 5, 10, 20, 30, 50, 75, 100],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
}

knn_clf = GridSearchCV(knn, knn_params, verbose=1, n_jobs=4, error_score=0)

knn_clf.fit(df[x_features],df['output'])

print('The best estimator was {}, with {:.3f}% accuracy'.format(knn_clf.best_estimator_, (knn_clf.best_score_ * 100)))
```
