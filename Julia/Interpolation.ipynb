```julia
using Plots
```

Define a linear function:

```julia
function FHV(v0, alpha, t)
    
    #=
    return an array with values of velocity as time increases
    
    v0    = float
    alpha = float
    t     = float
    =#
    
    [v0 + alpha * (t + i) for i in 0:25]
end
```

```julia
plot([i for i in 5:30], FHV(2, 0.5, 5))
```

Define the linear interpolation formula:

```julia
function linear_interpolate(p, p1, p2)
    p[2] = p1[2] + (p[1] - p1[1]) * (p2[2] - p1[2]) / (p2[1] - p1[1])
end
```

Get the values of t and the function outputs for then:

```julia
for i in 1:26
    print("t = ", i+4 ," --> ", FHV(2, 0.5, 5)[i], "\n")
end
```

We'll try to guess the value at t=10

```julia
linear_interpolate([10,0],[5,4.5],[15,9.5])
```

```julia
linear_interpolate([10,0],[7,5.5],[13,8.5])
```

```julia
linear_interpolate([10.,0],[9,6.5],[11,7.5])
```

Linear functions are pretty easy to guess

Now we'll try with a quadratic function:

```julia
function FHP(s0, v0, t, alpha)
    
    #=
    return an array with values of position as time increases
    
    s0    = float
    v0    = flaot
    alpha = float
    t     = float
    =#
    
    [s0 + (v0 * t + i) + alpha * ((t + i) ^ 2 / 2) for i in 0:25]
end
```

Get the values of t and the function outputs for then in this new function:

```julia
for i in 1:26
    print("t = ", i+2 ," --> ", FHP(5, 10, 3, 0.5)[i], "\n")
end
```

```julia
plot([i for i in 3:28], FHP(5, 10, 3, 0.5))
```

Define the quadratic interpolation function:

```julia
function quadratic_interpolate(p, p0, p1, p2)
    a = ((p[1] - p1[1]) * (p[1] - p2[1])) / ((p0[1] - p1[1]) * (p0[1] - p2[1])) * p0[2]
    b = ((p[1] - p0[1]) * (p[1] - p2[1])) / ((p1[1] - p0[1]) * (p1[1] - p2[1])) * p1[2]
    c = ((p[1] - p0[1]) * (p[1] - p1[1])) / ((p2[1] - p0[1]) * (p2[1] - p1[1])) * p2[2]
    p[1] = a + b + c
end
```

Let's try to guess the value at t=10 again:

```julia
quadratic_interpolate([10.,0.], [3.,37.25], [17.,121.25], [24.,200.])
```

```julia
quadratic_interpolate([10.,0.], [3.,37.25], [15.,103.25], [17.,121.25])
```
